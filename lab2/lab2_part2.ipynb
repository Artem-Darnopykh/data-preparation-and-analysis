Необхідно виконати завдання, використовуючи pandas dataframe, проаналізувати часові витрати на виконання процедур (профілювання часу виконання здійснити за допомогою модуля timeit).

# Завдання 1:
Звантажити та відкрити (вручну або через запропонований скрипт на сайті) наступний датасет: Individual Household Electric Power Consumption Dataset

```python
import pandas as pd
import numpy as np
import os
import urllib.request
import zipfile
import timeit

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00235/household_power_consumption.zip"
zip_path = "household_power_consumption.zip"
txt_file = "household_power_consumption.txt"

if not os.path.exists(txt_file):
    print("Файл не знайдено. Початок завантаження архіву...")
    urllib.request.urlretrieve(url, zip_path)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(".")
    print("Архів завантажено та розпаковано успішно.")
else:
    print("Файл вже існує.")

start_time = timeit.default_timer()
df = pd.read_csv(txt_file, sep=';', na_values=['?'], low_memory=False)
end_time = timeit.default_timer()

print(f"Початкова кількість рядків: {len(df)}")
print(f"Час зчитування файлу: {end_time - start_time:.4f} сек.")

df.head(11)
```

# Завдання 2:
Здійснити data cleaning

```python
numeric_columns = [
    'Global_active_power', 'Global_reactive_power', 'Voltage', 
    'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3'
]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Тепер рахуємо всі виявлені пропуски
missing_before = df.isna().sum().sum()
print(f"Знайдено порожніх клітинок (NaN) після обробки символів: {missing_before}")

# Вимірюємо час безпосередньо видалення рядків
start_time = timeit.default_timer()
df_cleaned = df.dropna()
end_time = timeit.default_timer()

print(f"Залишилося рядків: {len(df_cleaned)}")
print(f"Видалено рядки: {len(df) - len(df_cleaned)}")
print(f"Час виконання очищення: {end_time - start_time:.4f} сек.")

print("\nТипи даних після конвертації:")
print(df_cleaned[numeric_columns].dtypes)

df = df_cleaned
```

# Завдання 3:
Обрати всі записи, у яких загальна активна споживана потужність перевищує 5 кВт.

```python
def filter_high_active_power(df):
    return df[df['Global_active_power'] > 5]

start_time = timeit.default_timer()
high_power_df = filter_high_active_power(df)
end_time = timeit.default_timer()

print(f"Кількість записів з потужністю > 5 кВт: {len(high_power_df)}")
print(f"Час виконання вибірки: {end_time - start_time:.6f} сек.")

high_power_df.head(10)
```

# Завдання 4:
Обрати всі записи, у яких сила струму лежить в межах 19-20 А, для них виявити ті, у яких пральна машина та холодильних споживають більше, ніж бойлер та кондиціонер.

```python
def filter_by_intensity_and_consumption(df):
    condition = (df['Global_intensity'] >= 19) & \
                (df['Global_intensity'] <= 20) & \
                (df['Sub_metering_2'] > df['Sub_metering_3'])
    return df[condition]

start_time = timeit.default_timer()
filtered_subset = filter_by_intensity_and_consumption(df)
execution_time = timeit.default_timer() - start_time

print(f"Кількість записів, що відповідають умовам: {len(filtered_subset)}")
print(f"Час виконання процедури: {execution_time:.6f} сек.")

filtered_subset.head()
```

# Завдання 5:
Обрати випадковим чином 500000 записів (без повторів елементів вибірки), для них обчислити середні величини усіх 3-х груп споживання електричної енергії

```python
def random_sampling_analysis(df, n=500000):
    sample_df = df.sample(n=n, replace=False)
    means = sample_df[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean()
    return means

start_time = timeit.default_timer()
means_result = random_sampling_analysis(df)
execution_time = timeit.default_timer() - start_time

print("Середні величини усіх 3-х груп споживання електричної енергії:")
print(means_result.astype(str) + " Вт-год")
print(f"Час виконання процедури: {execution_time:.6f} сек.")
```

# Завдання 6:
Обрати ті записи, які після 18-00 споживають понад 6 кВт за хвилину в середньому, серед відібраних визначити ті, у яких основне споживання електроенергії у вказаний проміжок часу припадає на пральну машину, сушарку, холодильник та освітлення (група 2 є найбільшою), а потім обрати кожен третій результат із першої половини та кожен четвертий результат із другої половини.


```python
def complex_filtration(df):
    if 'Datetime' not in df.columns:
        df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], dayfirst=True)
    
    subset = df[(df['Datetime'].dt.hour >= 18) & (df['Global_active_power'] > 6)]
    
    subset = subset[(subset['Sub_metering_2'] > subset['Sub_metering_1']) & 
                    (subset['Sub_metering_2'] > subset['Sub_metering_3'])]
    
    n = len(subset)
    half = n // 2
    
    first_half = subset.iloc[:half:3]
    second_half = subset.iloc[half::4]
    
    return pd.concat([first_half, second_half])

start_time = timeit.default_timer()
final_result = complex_filtration(df)
execution_time = timeit.default_timer() - start_time

print(f"Кількість отриманих записів: {len(final_result)}")
print(f"Час виконання процедури: {execution_time:.6f} сек.")
final_result.head(5)
```

# Завдання 7:
Пронормувати та стандартизувати вибраний датасет


```python
def normalize_and_standardize(df):
    numeric_cols = [
        'Global_active_power', 'Global_reactive_power', 'Voltage', 
        'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3'
    ]
    
    df_norm = df.copy()
    for col in numeric_cols:
        df_norm[col] = (df_norm[col] - df_norm[col].min()) / (df_norm[col].max() - df_norm[col].min())
    
    df_stand = df.copy()
    for col in numeric_cols:
        df_stand[col] = (df_stand[col] - df_stand[col].mean()) / df_stand[col].std()
        
    return df_norm, df_stand

start_time = timeit.default_timer()
df_normalized, df_standardized = normalize_and_standardize(df)
execution_time = timeit.default_timer() - start_time

print(f"Час виконання процедури: {execution_time:.6f} сек.")

print("\nПерші рядки нормованого датасету:")
print(df_normalized.head())

print("\nПерші рядки стандартизованого датасету:")
print(df_standardized.head())
```

# Завдання 8:
Підрахувати коефіцієнт Пірсона та Спірмена для двох integer/real атрибутів.

```python
def calculate_correlations(df, attr1, attr2):
    # Примусово конвертуємо в числа, щоб уникнути TypeError
    s1 = pd.to_numeric(df[attr1], errors='coerce')
    s2 = pd.to_numeric(df[attr2], errors='coerce')
    
    pearson_corr = s1.corr(s2, method='pearson')
    try:
        spearman_corr = s1.corr(s2, method='spearman')
    except Exception:
        spearman_corr = "Помилка: встановіть scipy (pip install scipy)"
        
    return pearson_corr, spearman_corr

attr1 = 'Global_active_power'
attr2 = 'Global_intensity'

start_time = timeit.default_timer()
p_corr, s_corr = calculate_correlations(df, attr1, attr2)
execution_time = timeit.default_timer() - start_time

print(f"Коефіцієнт Пірсона: {p_corr:.6f}")
print(f"Коефіцієнт Спірмена: {s_corr}")
print(f"Час виконання: {execution_time:.6f} сек.")
```

# Завдання 9:
Провести One Hot Encoding категоріального атрибута.


```python
def apply_one_hot_encoding(df):
    if 'Datetime' not in df.columns:
        df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], dayfirst=True)
    
    def get_period(hour):
        if 6 <= hour < 12: return 'Morning'
        elif 12 <= hour < 18: return 'Day'
        elif 18 <= hour < 24: return 'Evening'
        else: return 'Night'

    df_sample = df.sample(10000).copy()
    df_sample['Day_Period'] = df_sample['Datetime'].dt.hour.apply(get_period)
    
    encoded_df = pd.get_dummies(df_sample, columns=['Day_Period'], prefix='Period')
    return encoded_df

start_time = timeit.default_timer()
df_encoded = apply_one_hot_encoding(df)
execution_time = timeit.default_timer() - start_time

print(f"Час виконання One Hot Encoding: {execution_time:.6f} сек.")
print("Нові стовпці після кодування:")
print([col for col in df_encoded.columns if 'Period_' in col])

df_encoded.head()
```
